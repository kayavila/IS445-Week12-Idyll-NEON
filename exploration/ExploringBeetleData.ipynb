## Exploring the Sorted Field Data 

```python
fielddata = pd.read_csv(data_dir + '/bet_fielddata.csv')

# Look at just bout eventID MOAB.2018.16 at plot MOAB_010
fielddata[(fielddata['eventID'] == 'MOAB.2018.16') & (fielddata['plotID'] == 'MOAB_010')].T
```

```python
sorting = pd.read_csv(data_dir + '/bet_sorting.csv',
                     usecols=['plotID', 'trapID', 'processingDate', 'sampleType', 'taxonID', 'taxonRank', 
                              'individualCount', 'nativeStatusCode', 'scientificName'])
```

```python
# What kind of bycatch is there?
sorting['sampleType'].value_counts()
```

```python
# Show only ground beetles from plot MOAB_010
sorting[(sorting['plotID'] == 'MOAB_010') & (sorting['sampleType'] == 'carabid')].head(10)
```

```python
sorting.count()
```

```python
# Look at null values
sorting.isnull().sum()
```

```python
# Number of rows with null values for taxonID is equal to sampleType "invert bycatch"
# What are the ones that are not carabid but not invert bycatch?
sorting[(sorting['sampleType'] != 'carabid') & (sorting['taxonID'].notna())]['scientificName'].value_counts()
```

```python
# How many different kinds of ground beetles were collected?
print('{} unique beetle species:'.format(sorting[sorting['sampleType'] == 'carabid']['scientificName'].nunique()))

# What species?
sorting[sorting['sampleType'] == 'carabid']['scientificName'].value_counts().head(100)
```
